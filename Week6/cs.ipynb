```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = 100
pd.options.display.max_rows = 100
```

```python
data = pd.read_csv('data.csv')
```

```python
data.shape
```

```python
import copy
df = copy.deepcopy(data)
```

```python
df.head()
```

```python
list(df.columns)
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('-','_')
list(df.columns)
```

```python
df.head()
```

```python
df['bankrupt?'].value_counts()
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
plt.figure(figsize=(10, 80))
sns.heatmap(df.corr().iloc[:,[0]].sort_values(by='bankrupt?', ascending=False),annot=True)
plt.show()
```

```python
df.nunique().sort_values(ascending=True)
```

```python
df.drop(columns=['liability_assets_flag', 'net_income_flag'], axis=1, inplace=True)
```

```python
X = df.drop(columns=['bankrupt?'], axis=1)
y = df['bankrupt?']
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                        test_size=0.3,
                                                        random_state=42,
                                                        shuffle=True, 
                                                        stratify=y)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaler = scaler.fit(X_train)

x_train_scaled = scaler.transform(X_train)
x_test_scaled = scaler.transform(X_test)
```

```python
x_train_scaled.shape, x_test_scaled.shape, y_train.shape, y_test.shape
```

```python
X_train = copy.deepcopy(x_train_scaled)
X_test = copy.deepcopy(x_test_scaled)
```

### Bagging Classifier

Decision Tree

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
bagging_dt_model = BaggingClassifier(estimator=DecisionTreeClassifier(random_state=42),
                                     n_estimators=150, 
                                     max_samples=0.5, 
                                     max_features=8,
                                     bootstrap=True,
                                     bootstrap_features=True,
                                     random_state=101)
```

```python
bagging_dt_model.fit(X_train, y_train)
```

```python
bagging_dt_model.score(X_test, y_test)
```

```python
bagging_dt_model_pred = bagging_dt_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, bagging_dt_model_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report

print("Confusion matrix : \n", confusion_matrix(y_true=y_test, y_pred=bagging_dt_model_pred))

print("Classification report : \n", classification_report(y_true=y_test, y_pred=bagging_dt_model_pred))

print("Accuracy score : ", accuracy_score(y_true=y_test, y_pred=bagging_dt_model_pred))
print("Precision score : ", precision_score(y_true=y_test, y_pred=bagging_dt_model_pred))
print("Recall score : ", recall_score(y_true=y_test, y_pred=bagging_dt_model_pred))
print("F1 score : ", f1_score(y_true=y_test, y_pred=bagging_dt_model_pred))

# TN FP
# FN TP
# 0  1
```

LightGBM

```python
import lightgbm as lgb
```

```python
lgb_params = {
    'learning_rate': 0.03,
    'num_leaves': 24,
    'max_depth': 12,
    'num_iterations': 1000,
    'scale_pos_weight': df['bankrupt?'].value_counts()[0] / df['bankrupt?'].value_counts()[1]
    }
```

```python
lgb_model = lgb.LGBMClassifier(**lgb_params)
```

```python
lgb_model.fit(X_train, y_train, eval_set=[(X_test, y_test)])
```

```python
lgb_score = lgb_model.score(X_test, y_test)
lgb_score
```

```python
lgb_model_pred = lgb_model.predict(X_test)
```

```python
print("Confusion matrix : \n", confusion_matrix(y_true=y_test, y_pred=lgb_model_pred))

print("Classification report : \n", classification_report(y_true=y_test, y_pred=lgb_model_pred))

print("Accuracy score : ", accuracy_score(y_true=y_test, y_pred=lgb_model_pred))
print("Precision score : ", precision_score(y_true=y_test, y_pred=lgb_model_pred))
print("Recall score : ", recall_score(y_true=y_test, y_pred=lgb_model_pred))
print("F1 score : ", f1_score(y_true=y_test, y_pred=lgb_model_pred))

# TN FP
# FN TP
# 0  1
```
