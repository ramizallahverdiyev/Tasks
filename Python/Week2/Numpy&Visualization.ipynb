Question 8

```python
array = np.array(["Hello", "World", "PyThOn"])

capitalized = np.char.capitalize(array)
lowercased = np.char.lower(array)
uppercased = np.char.upper(array)
swapcased = np.char.swapcase(array)
titlecased = np.char.title(array)

print("Original Array:", array)
print("Capitalized:", capitalized)
print("Lowercased:", lowercased)
print("Uppercased:", uppercased)
print("Swapcased:", swapcased)
print("Titlecased:", titlecased)
```

```python
df = pd.read_csv('cars.csv')
df.head(5)
```

Question 1

a.

```python
highway_mpg = df['hwy']
city_mpg = df['cty']

plt.scatter(highway_mpg, city_mpg, alpha=0.7, color='blue')

plt.xlabel('Highway MPG')
plt.ylabel('City MPG')
plt.title('Relationship between Highway and City MPG')

plt.show()
```

b. Positive Correlation

Question 2

```python
import seaborn as sns

sns.pairplot(df, hue='drv', diag_kind='kde', palette='Set2')

plt.show()
```

Question 3

```python
sorted_ = df['class'].value_counts().index

sns.countplot(data=df, x='class', order=sorted_,palette='Set2')

plt.xlabel('Car Type')
plt.ylabel('Count')
plt.title('Frequency of Car Types')

plt.show()
```

Question 4

```python
sns.barplot(data=df, x='drv', y='cyl', palette='Set2')

plt.xlabel('Drivetrain')
plt.ylabel('Number of Cylinders')
plt.title('Number of Cylinders for Each Drivetrain')

# Show the plot
plt.show()
```

r is most  frequent.

Question 5

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='class', y='displ', palette='Set2')
plt.xlabel('Car Class')
plt.ylabel('Engine Displacement')
plt.title('Engine Displacement by Car Class (Boxplot)')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='class', y='displ', palette='Set2')
plt.xlabel('Car Class')
plt.ylabel('Engine Displacement')
plt.title('Engine Displacement by Car Class (Violinplot)')
plt.xticks(rotation=45)
plt.show()
```
